```python
# Importing important modules for compute the fields
import sympy                                    # import the symbolic library
from sympy.abc import x, y                      # in all code x and y are treated as symbolics variables
from sympy.utilities.lambdify import lambdify   # import function that allows us to transform symbolics into numerical functions
import numpy as np                              # numpy for numerical calculus <3
import matplotlib.pyplot as plt                 # library for 

xc = 0       # the "wire" distance between the origin and x axis direction                             
yc = 5       # the "wire" distance between the origin and y axis direction
```

```python
#definition of potential function of two charges ----- Symbolic function
pot_sym = 1./((x-xc)**2+(y-yc)**2)**0.5 - 1./((x-xc)**2+(y+yc)**2)**0.5  # equipotential of two wires ( the real and its image)
```

```python
#definition of electric fild by -Grad(potential) ---- Symbolic function
eletric_field = [-sympy.diff(pot_sym,x),-sympy.diff(pot_sym,y)]          # field of two "wires" (the real and its image)
```

```python
# Define the "wire's" electric fild and its potencial in numerical way
ex = lambdify((x,y),eletric_field[0],'numpy')     #electric field in x direction
ey = lambdify((x,y),eletric_field[1],'numpy')     #electric field in y direction
pot_num = lambdify((x,y),pot_sym,'numpy')         #potential function
```

### Once the code upward are working lets implement the lines necessary to the plot of equipotentials and the streamlines ( reprensents the field lines)

```python
# create a mesh grid directaly 
xx,yy=np.meshgrid(np.linspace(-20,20,200),np.linspace(-20,20,200))
```

```python
# The necessary lines for plotting the entire graph
fig, ax = plt.subplots(figsize=(8,8))                     # The graph's lenght
ax.streamplot(xx,yy,ex(xx,yy),ey(xx,yy),density=2)        # The line forces
plt.contour(xx,yy,pot_num(xx,yy),550,colors='black')      # The equipotential lines
plt.xlim(-13,13)                                          # limit of the x label in the graph
plt.ylim(-13,13)                                          # Limit of the y label in the graph
plt.grid(True)                                            # put the grid for better vizualization
plt.show()                                                # show the graph
```
